# 7. Univariate Analysis

## Functions for Plotting Configurations

```python
def plot_histogram(dataframe, column, bins=10):
    # Create the histogram and capture the bin edges
    counts, bin_edges, patches = plt.hist(
        dataframe[column], 
        bins=bins,  # Adjust the number of bins as needed
        color="#5E686D", 
        edgecolor="black"
    )

    # Format the column name for the title
    title = column.replace('_', ' ').title() + " Distribution"

     # Format the label for the x-axis
    xlabel = column.replace('_', ' ').title()

    # Plot the histogram using Pandas for additional customization
    ax = dataframe[column].plot(
        kind="hist",
        bins=bins,
        title=title,
        xlabel=xlabel,
        color="#5E686D",
        edgecolor="black"
    )

    # Set x-axis ticks to match bin edges
    ax.set_xticks(bin_edges)
    ax.set_xticklabels([f'{int(edge)}' for edge in bin_edges])

    # Annotate the y-axis with the count for each bar
    for bar in ax.patches:
        height = bar.get_height()
        x = bar.get_x() + bar.get_width() / 2
        ax.annotate(f'{int(height)}', xy=(x, height), xytext=(0, 5),
                    textcoords='offset points', ha='center', va='bottom')

    return plt, ax  # Return the plt object and ax (to use later for saving)

def plot_bar(dataframe, column):

    # Format the column name for the title
    title = column.replace('_', ' ').title() + " Distribution"

    # Format the label for the x-axis
    xlabel = column.replace('_', ' ').title()
    
    # Create the bar plot
    ax = dataframe[column].value_counts().plot(
        kind="bar", 
        title=title, 
        xlabel=xlabel, 
        ylabel="Frequency", 
        color="#5E686D", 
        edgecolor="black"
    )

    # Annotate each bar with the exact value (y-axis)
    for bar in ax.patches:
        height = bar.get_height()  # Get the height (y-value) of each bar
        ax.annotate(f'{int(height)}',  # Format the annotation
                    xy=(bar.get_x() + bar.get_width() / 2, height),  # Position of the annotation
                    xytext=(0, 5),  # Offset the annotation slightly above the bar
                    textcoords='offset points',
                    ha='center',  # Center align the text horizontally
                    va='bottom')  # Align text above the bar

    return plt, ax  # Return the plt object and ax for further customization or saving

def plot_pie(dataframe, column):
    # Format the column name for the title
    title = column.replace('_', ' ').title() + " Distribution"
    
    # Calculate the value counts and generate the pie chart
    ax = dataframe[column].value_counts().plot(
        kind="pie", 
        title=title,  # Use the formatted title
        autopct="%1.1f%%",  # Display percentages
        ylabel=""  # Remove the default y-label for a cleaner look
    )
    
    return plt, ax  # Return the plt object and ax for further customization or saving

def plot_grouped_bar_chart(data, title, xlabel, ylabel, rot=0):
    # Create the bar plot
    ax = data.plot(
        kind="bar",
        title=title, 
        xlabel=xlabel, 
        ylabel=ylabel, 
        color="#5E686D", 
        edgecolor="black",
        rot=rot
    )

    # Annotate each bar with the exact value
    for bar in ax.patches:
        height = bar.get_height()  # Get the height (y-value) of each bar
        ax.annotate(f'{height:.2f}',  # Format the annotation to 2 decimal places
                    xy=(bar.get_x() + bar.get_width() / 2, height),  # Position of the annotation
                    xytext=(0, 5),  # Offset the annotation slightly above the bar
                    textcoords='offset points',
                    ha='center',  # Center align the text horizontally
                    va='bottom')  # Align text above the bar

    return plt, ax  # Return plt and ax for further customization or saving
    
def plot_scatter(dataframe, x, y, color_column, colormap="viridis", title=None, xlabel=None, ylabel=None, center_point=None, center_label=None):
    # Create the scatter plot
    ax = dataframe.plot(
        kind="scatter",
        x=x,
        y=y,
        c=color_column,
        colormap=colormap,
        title=title if title else f"{y} vs. {x}",
        xlabel=xlabel if xlabel else x,
        ylabel=ylabel if ylabel else y
    )

    # Add a marker for the center point if provided
    if center_point:
        ax.scatter(
            center_point[0],  # Longitude (x)
            center_point[1],  # Latitude (y)
            color="white",  # Make it more visible against the colormap
            edgecolor="black",  # Add a border for contrast
            s=150,  # Larger size to make it more prominent
            label=center_label if center_label else "Center Point"
        )
        ax.legend()

    return plt, ax  # Return for further customization or saving



def save_plot(plt, filename, folder='plots'):
    # Save the plot in the specified folder
    plt.savefig(f'{folder}/{filename}.png', dpi=300, bbox_inches='tight')
```

## 7.7. License

```python
def categorize_license_status(license_status):
    if license_status == "Unknown":
        return "Unknown"

    return "Licensed"
    
# Categorize license status and generate the pie chart
listings["license_status"] = listings["license"].apply(categorize_license_status)

plt, ax = plot_pie(listings, "license_status")

# Save the plot if desired
save_plot(plt, "license_status_distribution", folder='plots')

# Display the plot
plt.show()
```

# 8. Multivariate Analysis

## 8.1. Correlation Matrix

```python
listings_numerical_columns = listings.select_dtypes(include=[int, float]).drop(columns=["id", "host_id", "longitude", "latitude"])

corr_matrix = listings_numerical_columns.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

## 8.2 Group By Analysis

### 8.2.1 Average Price of Each Neighbourhood Group

```python
# Group data by neighbourhood_group and calculate the mean price
listings_grouped_by_ng = listings.groupby("neighbourhood_group", observed=True)
avg_price_by_ng = listings_grouped_by_ng["price"].mean().sort_values()

# Use the function to create the bar plot
plt, ax = plot_grouped_bar_chart(
    avg_price_by_ng,
    title="Average Price on Each Neighbourhoods",
    xlabel="Neighbourhoods",
    ylabel="Average Price",
    rot=90
)

# Save the plot
save_plot(plt, "average_price_neighbourhoods", folder="plots")

# Show the plot
plt.show()
```

### 8.2.2 Average Price of Each Room Type

```python
# Group data by room_type and calculate the mean price
listings_grouped_by_rt = listings.groupby("room_type", observed=True)
avg_price_by_rt = listings_grouped_by_rt["price"].mean()

# Use the function to create the bar plot
plt, ax = plot_grouped_bar_chart(
    avg_price_by_rt,
    title="Average Price Of Each Room Type",
    xlabel="Room Type",
    ylabel="Average Price",
    rot=0  # Set rotation for x-axis labels to 0
)

# Save the plot
save_plot(plt, "average_price_room_type", folder="plots")

# Show the plot
plt.show()
```

# 9. Geospatial Data Visualizations

```python
# Use the function to plot the scatter chart with Berlin's center highlighted
plt, ax = plot_scatter(
    dataframe=listings,
    x="longitude",
    y="latitude",
    color_column="price",
    colormap="RdYlGn_r",
    title="Price (per Night) Based on the Location",
    xlabel="Longitude",
    ylabel="Latitude",
    center_point=(13.4050, 52.5200),  # Longitude, Latitude of Berlin's center
    center_label="Berlin Center"
)

# Save the plot
save_plot(plt, "price_location_scatter", folder="plots")

# Show the plot
plt.show()
```

```python
# Neighborhoods in the specified order
areas = [
    'Mitte', 'Friedrichshain-Kreuzberg', 'Pankow', 'Charlottenburg-Wilm.',
    'Neukölln', 'Tempelhof - Schöneberg', 'Treptow - Köpenick', 
    'Steglitz - Zehlendorf', 'Lichtenberg', 'Reinickendorf', 
    'Spandau', 'Marzahn - Hellersdorf'
]

# Approximate longitudes and latitudes for the neighborhoods
longitudes = [
    13.4050, 13.4410, 13.4285, 13.3060, 13.4350, 
    13.3622, 13.5738, 13.2529, 13.4997, 13.2912, 
    13.2003, 13.5652
]
latitudes = [
    52.5200, 52.5158, 52.5695, 52.5208, 52.4720, 
    52.4713, 52.4501, 52.4550, 52.5123, 52.5925, 
    52.5369, 52.5328
]

# Actual counts
counts = [1835, 1722, 1330, 934, 731, 639, 412, 263, 238, 162, 131, 99]

# Coordinates for Berlin city center
berlin_center = {"latitude": 52.5200, "longitude": 13.4050}

# Create scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(longitudes, latitudes, color='blue', marker='o', s=100, label='Neighborhoods')

# Add Berlin city center point
plt.scatter(berlin_center["longitude"], berlin_center["latitude"], color='red', marker='x', s=150, label='Berlin Center')

# Annotate each point with neighbourhood name and count
for i, area in enumerate(areas):
    label = f"{area} ({counts[i]})"
    plt.text(longitudes[i], latitudes[i] + 0.005, label, fontsize=9, ha='center')  # Move text slightly higher

# Add labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Berlin Neighborhoods with Listing Counts')
plt.grid(True)
plt.legend()

# Save the plot as a PNG file
plt.savefig('berlin_neighborhoods_with_listings_counts_plot.png', dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```
